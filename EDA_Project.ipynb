```python
#Box Plot of Weather Delay
sns.boxplot(x='Arr_delay',data=df)  
```

```python
#Box Plot of Weather Delay
sns.boxplot(x='Year',y='Weather_delay',data=df)  
```

Heat map of Correlation plot

```python

plt.figure(figsize=(5,5))
sns.heatmap(df.corr(),cmap='BrBG')
plt.title('Heat Map')
plt.show
```

Histogram

```python
#Histogram of Security Delay
plt.figure(figsize=(20,15))
sns.histplot(x='Year',y='Security_delay',data=df,cmap='pink')
plt.title('Histogram')
plt.xlabel('Year')
plt.ylabel('Security_delay')
plt.show
```

```python
#Histogram of Carrier Delay

plt.figure(figsize=(20,15))
sns.histplot(x='Year',y='Carrier_delay',data=df,cmap='pink')
plt.title('Histogram')
plt.xlabel('Year')
plt.ylabel('Carrier_delay')
plt.show
```

Bar Plot

```python
#Barplot of Arrival Delay per Year

sns.barplot(x='Year',y="Arr_delay",data=df)
plt.title('Arrival Delay per Year')
plt.xlabel('Year')
plt.ylabel('Delay')
plt.show
```

Line Plot

```python
#Lineplot of Arrival Delay per Year

sns.lineplot(x='Year',y="Arr_delay",data=df)
plt.title('Arrival Delay per Year')
plt.xlabel('Year')
plt.ylabel('Delay')
plt.show
```

Scatter Plot

```python
##Scatter Plot of Arrival Delay per Year

plt.scatter(df['Year'],df['Arr_delay'],alpha=0.5)
plt.title('Scatter Plot of Arrival Delay')
plt.xlabel('Year')
plt.ylabel('Delay')
plt.show
```
